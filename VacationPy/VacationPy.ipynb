```python
# File to Load
file_to_load = '../WeatherPy/output_data/cities.csv'

# Read Purchasing File and store into Pandas data frame
heatmap_data_df = pd.read_csv(file_to_load)

# View Heatmap Data Frame
heatmap_data_df.style.hide_index()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Display Google Maps
fig = gmaps.figure()

fig
```

```python
# Create a Humidity Heatmap layer
# Store 'Lat' and 'Lng' into  locations 
locations = heatmap_data_df[['Lat','Lng']].astype(float)

# Store humidity into weights
humidity = heatmap_data_df['Humidity']


heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=150,
                                 point_radius = 2)

fig.add_layer(heat_layer)

fig
```
